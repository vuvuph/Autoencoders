```python
#Load main IoT-23 dataset
main_df = pd.read_csv("/content/drive/MyDrive/IoT-23 Dataset/iot23_combined.csv")

#Load remaining datasets
df17 = pd.read_csv("/content/drive/MyDrive/IoT-23 Dataset/iot23_combined_df17.csv")
df33 = pd.read_csv("/content/drive/MyDrive/IoT-23 Dataset/iot23_combined_df33.csv")
df39 = pd.read_csv("/content/drive/MyDrive/IoT-23 Dataset/iot23_combined_df39.csv")
df43 = pd.read_csv("/content/drive/MyDrive/IoT-23 Dataset/iot23_combined_df43.csv")
```

```python
#Combine datasets to one and save to .csv file
main_df = pd.concat([main_df, df17, df33, df39, df43], ignore_index=True)

main_df.to_csv("/content/drive/MyDrive/IoT-23 Dataset/iot23_combined_final.csv", index=False)
```

```python
pd.options.display.max_rows = 300
pd.options.display.max_columns = 300
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
df = pd.get_dummies(df, columns=['proto'])
df = pd.get_dummies(df, columns=['conn_state'])
```

```python
df
```

```python
df['duration'] = df['duration'].astype(str).str.replace('-','0')
df['orig_bytes'] = df['orig_bytes'].astype(str).str.replace('-','0')
df['resp_bytes'] = df['resp_bytes'].astype(str).str.replace('-','0')
```

```python
df.fillna(-1,inplace=True)
```
